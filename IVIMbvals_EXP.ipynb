```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl
%matplotlib inline
import csv

import scipy.optimize as op
import scipy.stats
from scipy.optimize import curve_fit
import scipy.io
```

```python
def makehist(matf):
    (x,y,z)=matf.shape
    n = int(z/2)
    vals = []
    for i in range(0,x):
        for j in range(0,y):
            for k in range(n-2,n+2):
                vals.append(matf[i][j][k])
    return (np.array(vals)).astype(float)
                
    
```
